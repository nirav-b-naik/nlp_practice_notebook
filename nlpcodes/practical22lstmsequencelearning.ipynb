```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

<b><u> indexing each character from alphabets string

```python
list(enumerate(alphabets));
```

<b><u> creating character dict from alphabets string

```python
char_to_int = dict((c,i) for i,c in enumerate(alphabets))
```

```python
int_to_char = dict((i,c) for i,c in enumerate(alphabets))
```

<b><u> creating x and y (variable and respond)

```python
seq_length = 1

x = list()

y = list()

for i in range(len(alphabets)-seq_length):
    
    seq_in = alphabets[i:i+seq_length]
    
    seq_out = alphabets[i+seq_length]
    
    x.append([char_to_int[char] for char in seq_in])
    
    y.append(char_to_int[seq_out])
```

```python
import numpy as np
```

```python
x = np.reshape(x,(len(x),seq_length,1))
```

<b><u> feature scaling

```python
x = x /(len(alphabets)-1)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.shape
```

<b><u> SimpleRNN Model + LSTM Layer

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    
    Bidirectional(LSTM(32)),
    
    Dense(26,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss="categorical_crossentropy",
              metrics='accuracy')
```

```python
# history = model.fit(x,y,
#                     epochs=500,
#                    batch_size = 1)
```

```python
history1 = model.fit(x,y,
                    epochs=1000,
                   batch_size = 1)
```

```python
model.summary()
```

```python
model.evaluate(x,y,batch_size=1)
```

<b><u> Prediction on new data

```python
new ='A'
```

```python
# converting char to integer in list type

new = [char_to_int[new]]
```

```python
# converting list to numpy array

new = np.array(new)
```

```python
# reshaping new as per input shape of x

new = np.reshape(new,(len(new),seq_length,1))
```

```python
# feature scaling as per input X

new = new/(len(alphabets)-1)
```

```python
# predict new with trained model

y_pred = model.predict(new)
```

```python
# previus y_pred was in list with probabilities

# the argmax function will give index of maximum prbability, it will be the integer value for new y_pred

y_pred = np.argmax (y_pred)
```

```python
# converting y_pred integer to y_pred char

y_pred = int_to_char[y_pred]
```

```python
# y_pred char

y_pred
```
