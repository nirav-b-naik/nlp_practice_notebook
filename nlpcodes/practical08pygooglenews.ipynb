```python
from pygooglenews import GoogleNews
```

```python
gn= GoogleNews()
```

```python
s = gn.search('Priyanka Chopra')
```

```python
for entry in s['entries']:
  print(entry['title'])
```

```python
hd=gn.geo_headlines('Surat')
```

```python
for entry in hd['entries']:
  print(entry['title'])
```
