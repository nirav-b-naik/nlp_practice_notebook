```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sigmoid function or logistic function

def logistic(z):
    
    val = 1 / (1 + np.exp(-z))
    
    return val
```

```python
x = np.arange(-5,5,0.25)
```

```python
y= x*23 + 45.12
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,logistic(x))
```

```python
logistic(1.45)
```

```python
# tanh function

def tanh_func(z):
    
    val = (np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z))
    
    return val
```

```python
plt.plot(x,tanh_func(x))
plt.grid()
```

```python
# softmax function

def fn_softmax(z):
    
    val = (np.exp(z))/np.sum(np.exp(z))
    
    return val
```

```python

plt.plot(x,fn_softmax(x))
```

```python
fn_softmax(x)
```

```python
fn_softmax([4.1,4.9,1.3]),np.sum(fn_softmax([4.1,4.9,1.3]))
```

```python
np.argmax(fn_softmax([4.1,4.9,1.3]))
```
