```python
grammar = "NP: {<DT>?<JJ>*<NN>}"

sent = "It's a dangerous business, Frodo, going out your door."
```

```python
from nltk import RegexpParser
from nltk import pos_tag
```

```python
chunk_parser = RegexpParser(grammar)
tags = pos_tag(word_tokenize(sent))
tree = chunk_parser.parse(tags)
tree
```

```python
#lemmatization

from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
```

```python
wordlist=['players', 'worst', 'biggest', 'children', 'ate','is']

for word in wordlist:
    
    print(word,": (verb): ",wnl.lemmatize(word,'v')) # verb

    print(word,": (noun): ",wnl.lemmatize(word,'n')) # noun

    print(word,": (adjective): ",wnl.lemmatize(word,'a')) # adjective

    print(word,": (adverb): ",wnl.lemmatize(word,'r')) # adverb
```
