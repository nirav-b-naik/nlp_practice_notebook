```python
from nltk.corpus import wordnet
```

```python
synset = wordnet.synsets('good')
print("word and type: " + synset[0].name())
print("synonym: " + synset[0].lemmas()[0].name())
print("The meaning of the word: " + synset[0].definition())
print("Example: " + str(synset[0].examples()))
```

```python
synset = wordnet.synsets('fly')
print("word and type: " + synset[0].name())
print("synonym: " + synset[0].lemmas()[0].name())
print("The meaning of the word: " + synset[0].definition())
print("Example: " + str(synset[0].examples()))
```

```python
for i in range(len(synset)):
    print(synset[i].lemmas())
```

```python
synset = wordnet.synsets('frame')

for i in range(len(synset)):
    print("word and type: " + synset[i].name())
    print("synonym: " + synset[i].lemmas()[0].name())
    print("The meaning of the word: " + synset[i].definition())
    print("Example: " + str(synset[i].examples()))
    print("\n")
```

```python
syn = list()
ant = list()

for synset in wordnet.synsets('cloud'):
    
    for lemma in synset.lemmas():
        syn.append(lemma.name())
        
        if lemma.antonyms():
            ant.append(lemma.antonyms()[0].name())

print('Synonyms: ',set(syn))
print('Antonyms: ',set(ant))
```

```python
# POS Tagging of Indian Languages

from nltk.corpus import indian
```

```python
indian.fileids()
```

```python
for x in indian.fileids():
    
    print(x,'\t: ',len(indian.raw(x)))
```

```python
sents = indian.sents('hindi.pos')
```

```python
for sent in sents:
    print(sent)
```

```python
sents = indian.tagged_sents('hindi.pos')

sentb = indian.tagged_sents('bangla.pos')

sent_t = indian.tagged_sents('telugu.pos')
```

```python
for sent in sents:
    print(sent)
    for (i,j) in sent:
        print(i,j)
    #print(sent[0][0],sent[0][1])
```

```python
from nltk.tag import tnt
from nltk.tokenize import word_tokenize
```

```python
# create the object of tagger

tagger = tnt.TnT()

taggerb = tnt.TnT()

tagger_t = tnt.TnT()
```

```python
tagger.train(data=sents)

taggerb.train(data=sentb)

tagger_t.train(data=sent_t)
```
